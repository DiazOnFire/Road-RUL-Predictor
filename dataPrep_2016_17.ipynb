```python
#Getting addresses
import pandas as pd
import requests
import json
import os
import dotenv
from dotenv import load_dotenv
import geopy
```

```python
d2016.head(40)
```

```python
load_dotenv()
api_keyy=os.getenv("api_key1")
```

```python
d2016['Lat']=0
d2016['Lon']=0
```

```python
geolocator=geopy.geocoders.Bing(api_keyy)
```

```python
def latlong(addressdata):
    location=geolocator.geocode(addressdata)
    if hasattr(location,'latitude') and hasattr(location,'longitude'):
        return(location.latitude,location.longitude)
    else:
        return 0,0

for ind in d2016.index:
    gg=latlong(d2016['Pavement_Section_From'][ind])
    d2016["Lat"][ind],d2016['Lon'][ind]=gg
```

```python
d2016.head()
```

Yet again, lot of coordinates are wrong and some of these sections are present in the 2014 data file where coordinates were given so we will try to indentify common paths and then change the Lat Long

```python
import math
```

```python
for ind in d2016.index:
    for ind1 in d2014.index:
        if d2016['Pavement_Section_From'][ind]==(d2014['Pavement_Section_From'][ind1]) or d2016['Pavement_Section_From'][ind]==(d2014['Pavement_Section_To'][ind1]):
            if math.isnan(d2014['Latitude'][ind1])==False:
                d2016['Lat'][ind]=d2014['Latitude'][ind1]
                d2016['Lon'][ind]=d2014["Longitude"][ind1]
    ind1=0
```

```python
d2017['Lat']=0
d2017['Lon']=0

for ind in d2017.index:
    gg1=latlong(d2017['Pavement_Section_From'][ind])
    d2017["Lat"][ind],d2017['Lon'][ind]=gg1
    

for ind in d2017.index:
    for ind1 in d2014.index:
        if d2017['Pavement_Section_From'][ind]==(d2014['Pavement_Section_From'][ind1]) or d2017['Pavement_Section_From'][ind]==(d2014['Pavement_Section_To'][ind1]):
            if math.isnan(d2014['Latitude'][ind1])==False:
                d2017['Lat'][ind]=d2014['Latitude'][ind1]
                d2017['Lon'][ind]=d2014["Longitude"][ind1]
    ind1=0 
```

```python
d2016['Lat'].isin(d2014['Latitude']).value_counts()
```
