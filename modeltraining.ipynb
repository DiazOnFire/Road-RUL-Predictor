```python
import pandas as pd
```

```python
d2014=pd.read_csv("G:\ml notebooks\Road_RUL\datasets\datasets\d2014f.csv")
d2015=pd.read_csv("G:\ml notebooks\Road_RUL\datasets\datasets\d2015f.csv")
d2016=pd.read_csv("G:\ml notebooks\Road_RUL\datasets\datasets\d2016f.csv")
d2017=pd.read_csv("G:\ml notebooks\Road_RUL\datasets\datasets\d2017f.csv")
```

## Lets combine our datasets to make a single dataframe

```python
finaldata=pd.concat([d2014,d2015,d2016,d2017])
```

```python
finaldata.drop(['AirPressure','Unnamed: 0'],inplace=True,axis=1)
```

```python
finaldata
```

```python
finaldata['P_combine']=finaldata.apply(lambda row: sorted([row['Pavement_Section_From'],row['Pavement_Section_To']]),axis=1)
finaldata['P_combine']=finaldata['P_combine'].apply(lambda x: ', '.join(map(str,x)))
```

```python
finaldata['P_combine']
```

```python
#Now lets sort it so that we can see data of one street over 4 years properly
# Sort the DataFrame
finaldata = finaldata.sort_values(by=['P_combine', 'Year'], ascending=True)
finaldata.drop(['Pavement_Section_From','Pavement_Section_To'],axis=1,inplace=True)
```

```python
#as pavement sections is irrelevant data, it wont really effect the result, so we drop them
finaldata.drop(['P_combine'],axis=1,inplace=True)
finaldata.drop(['DIR'],axis=1,inplace=True)
```

## Choosing appropriate model and training for inference
Our dataset is ready for training and this is essentially a time series data as it is data encapsulating 4 years of data, now we need to see the most appropriate model to predict RUL

From exising research papers and numerous articles, there is no fixed value of pci below which we can say the road is now in need of maintenance but in most cases, any value below 50 is generally indicates the road is in bad condition so we can fixate on 50 for now, IF we had more data, it would be possible to get a more accurate value.

Using a Time Series model like ARIMA is probably not effective here due to the dataset being quite small and it has only data across 4 years for a particular street, so we can just use regressive models to predict when the PCI value will dip below 50 numerically

We are going to be using a LSTM Model to predict when the value of PCI will go below 50 based on the above dataset, it is quite an unorthodox approach but the data acquired and the requirement are quite distant and there is no direct way to get desired result, so we will have to compromise a bit on the quality of output
