## We now try to get weather and traffic data based on the Pavement Section From and To columns which can give us an approximate location


```python
addresses=d2015[["Pavement_Section_From","Pavement_Section_To"]].copy()
```

```python
addresses
```

Now we have a list of addresses from which we can get the latitude and longitude.

```python
from dotenv import load_dotenv
import os
def configure():
    load_dotenv()
    
```

```python
configure()
api_key_bing=os.getenv('api_key1')
```

```python
import geocoder
import geopy
```

```python
geolocator=geopy.geocoders.Bing(api_key_bing)
```

```python
addresses
```

```python
def latlong(addressdata):
    location=geolocator.geocode(addressdata)
    if hasattr(location,'latitude') and hasattr(location,'longitude'):
        return(location.latitude,location.longitude)
    else:
        return 0,0

addresses["Lat"]=0
addresses["Lon"]=0
for ind in addresses.index:
    gg=latlong(addresses['Pavement_Section_From'][ind])
    addresses["Lat"][ind],addresses['Lon'][ind]=gg
    
```

```python
addresses['Lon'].value_counts()[0]
#Just two occurences where Latitude and Longitude could not be found, so we can just remove those occurences
```

```python
addresses=addresses[addresses['Lat']!=0]
```
